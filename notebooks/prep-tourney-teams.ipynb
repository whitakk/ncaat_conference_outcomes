```python
import os
os.chdir("..")

import pandas as pd

from src.classes.constants import SEED_ORDER, DATA_PATH, MODEL_PATH
```

```python
df = pd.read_csv(f'{DATA_PATH}/tourney_outcomes.csv') # see get_trank_data.py
df.head()
```

```python
seasons = sorted(list(df.season.unique()))

incl_cols = ['season','team','seed','conf']
data = []
for s in seasons:
    season_df = df[df.season==s]
    regions = sorted(list(season_df.region.unique()))
    for r in regions: 
        region_df = season_df[season_df.region==r]
        seeds = sorted(list(region_df.seed.unique()))
        max_seed = max(seeds)
        for sd in SEED_ORDER: 
            # first four are excluded so this will be unique to each team
            team_info = list(region_df[region_df.seed==sd][incl_cols].values[0])

            data.append(team_info)
```

```python
pd.DataFrame(data, columns=incl_cols).to_csv(f'{MODEL_PATH}tourney_teams.csv', index=False)
```
